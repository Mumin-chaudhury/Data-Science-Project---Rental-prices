```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('apartment_for_rent_train.csv')

df.head()
```

```python
df = df.drop_duplicates()

df_cleaned = df.dropna()

Q1 = df_cleaned['Price'].quantile(0.25)
Q3 = df_cleaned['Price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_cleaned = df_cleaned[(df_cleaned['Price'] >= lower_bound) & (df_cleaned['Price'] <= upper_bound)]

df_cleaned.reset_index(drop=True, inplace=True)
```

```python
print(df)
```

```python
from sklearn.tree import DecisionTreeRegressor

discrete_features = ['Balcony', 'Furniture', 'Renovation', 'Children_are_welcome', 'Pets_allowed']

for col in discrete_features:
    df_cleaned[col] = df_cleaned[col].astype('category').cat.codes

X_discrete = df_cleaned[discrete_features]
y_discrete = df_cleaned['Price']

X_train_a, X_test_a, y_train_a, y_test_a = train_test_split(X_discrete, y_discrete, test_size=0.2, random_state=42)

tree_a = DecisionTreeRegressor(max_depth=5, random_state=42)
tree_a.fit(X_train_a, y_train_a)

feature_importances_a = pd.Series(tree_a.feature_importances_, index=discrete_features).sort_values(ascending=False)
print("Top discrete variables:\n", feature_importances_a)
```
