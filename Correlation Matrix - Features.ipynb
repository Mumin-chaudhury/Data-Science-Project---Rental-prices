```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('apartment_for_rent_train.csv')

df = df.drop_duplicates()

df['Duration'] = df['Duration'].astype(str).str.lower().map({'monthly': 0, 'daily': 1})

df['Number_of_rooms'] = pd.to_numeric(df['Number_of_rooms'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')  # re-ensure

df_corr = df.dropna(subset=['Number_of_rooms', 'Price', 'Duration'])

correlation_matrix = df_corr[['Number_of_rooms', 'Price', 'Duration']].corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Number of Rooms, Price, Duration')
plt.tight_layout()
plt.savefig("correlation_matrix.png")
plt.show()

print("Correlation Matrix:\n", correlation_matrix)
```
